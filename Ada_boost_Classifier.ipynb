```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics 
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/nagarjuna843/Adaboost/main/income.csv')
df.head()
```

```python
inputs=df.drop(['education','income'],axis='columns')
target=df['income']
```

```python
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
adaboost=AdaBoostClassifier(n_estimators=3,learning_rate=1.0)
model=adaboost.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
output=pd.DataFrame({'y_test':y_test,'y_pred':y_pred})
output
```

```python
print("Accuracy: ",metrics.accuracy_score(y_test,y_pred))
```

```python
feature_importance=pd.DataFrame(model.feature_importances_,inputs.columns,columns=['Importance'])
feature_importance.sort_values('Importance',ascending=False)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model=LogisticRegression()
adaboost_1=AdaBoostClassifier(n_estimators=100,base_estimator=logistic_model,learning_rate=1)
```

```python
model=adaboost_1.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
print("Accuracy: ",metrics.accuracy_score(y_test,y_pred))
```
